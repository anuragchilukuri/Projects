```python
import pandas as pd


import tweepy

consumer_key = ''
consumer_secret = ''
access_token_key = ''
access_token_secret = ''
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token_key, access_token_secret)
api=tweepy.API(auth)
```

```python
query = 'Trump'
count = 100
max_tweets = 5000
searched_tweets = []
last_id = -1
while len(searched_tweets) < max_tweets:
    count = max_tweets - len(searched_tweets)
    try:
        new_tweets = api.search(q=query, count=count, max_id=str(last_id - 1))
        if not new_tweets:
            break
        searched_tweets.extend(new_tweets)
        last_id = new_tweets[-1].id
    except tweepy.TweepError as e:
        # depending on TweepError.code, one may want to retry or wait
        # to keep things simple, we will give up on an error
        break
```

```python
trump_tweet_data = []

for tweet in searched_tweets: #Create a list of lists of the attributes of tweets
    k=[]
    if hasattr(tweet,'retweeted_status'):
        k = k+[tweet.author.screen_name,tweet.author.listed_count,tweet.author.followers_count,tweet.text,tweet.retweet_count,tweet.retweeted_status.user.screen_name,tweet.retweeted_status.user.listed_count,tweet.retweeted_status.user.followers_count]
    else:
        k = k+[tweet.author.screen_name,tweet.author.listed_count,tweet.author.followers_count,tweet.text,tweet.retweet_count,'',0,0]
    trump_tweet_data.append(k)
print trump_tweet_data[:10] 
```

```python
trump_tweet_data_df = pd.DataFrame(trump_tweet_data) #Convert it into a dataframe and publish a csv file
trump_tweet_data_df.columns = ['User','listed_count','followers_count','text','retweets','original_author','original_author_listed_count','original_author_followers_count']
trump_tweet_data_df.to_csv('data.csv',encoding='utf-8')
trump_tweet_data_df
```

```python
trump_tweet_data_df_1 = trump_tweet_data_df[['User','listed_count','followers_count','text','retweets']]
trump_tweet_data_df_2 = trump_tweet_data_df[['original_author','original_author_listed_count','original_author_followers_count','text','retweets']]
trump_tweet_data_df_2.columns = ['User','listed_count','followers_count','text','retweets']
trump_tweet_data_df_merged = pd.concat([trump_tweet_data_df_1,trump_tweet_data_df_2],axis=0)
trump_tweet_data_df_merged
```

```python
trump_tweet_data_df_grouped = trump_tweet_data_df_merged.groupby('User')[['listed_count','followers_count','retweets']].mean()
```

```python
listed_count_weight = 0.44/(0.44+0.40+0.28)  #Caclualte weights based on their importances
follower_count_weight = 0.40/(0.44+0.40+0.28)
retweet_count_weight = 0.28/(0.44+0.40+0.28)
```

```python
from sklearn import preprocessing #Scale the data and score it
trump_tweet_data_df_grouped['listed_count'] = preprocessing.scale(trump_tweet_data_df_grouped['listed_count'])
trump_tweet_data_df_grouped['followers_count'] = preprocessing.scale(trump_tweet_data_df_grouped['followers_count'])
trump_tweet_data_df_grouped['retweets'] = preprocessing.scale(trump_tweet_data_df_grouped['retweets'])
trump_tweet_data_df_grouped['User_Score'] = trump_tweet_data_df_grouped['listed_count']*listed_count_weight+trump_tweet_data_df_grouped['followers_count']*follower_count_weight+trump_tweet_data_df_grouped['retweets']*retweet_count_weight
```

```python
trump_tweets_sorted = trump_tweet_data_df_grouped.sort('User_Score',ascending=False)
```
